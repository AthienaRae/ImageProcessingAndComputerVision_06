```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread("depth.jpeg")

if img is None:
    print("Image not found")
    exit()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

depth = cv2.normalize(gray, None, 0, 255, cv2.NORM_MINMAX)

depth_map = cv2.applyColorMap(depth, cv2.COLORMAP_JET)

cv2_imshow(img)
cv2_imshow(gray)
cv2_imshow(depth_map)
```
