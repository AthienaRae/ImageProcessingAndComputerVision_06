```python
import cv2, numpy as np
import matplotlib.pyplot as plt

img = cv2.cvtColor(cv2.imread("flower.jpg"), cv2.COLOR_BGR2RGB)
r, c = img.shape[:2]

resize = cv2.resize(img, (300,300))

M1 = np.float32([[1,0,50],[0,1,30]])
translate = cv2.warpAffine(img, M1, (c,r))

M2 = cv2.getRotationMatrix2D((c//2,r//2), 45, 1)
rotate = cv2.warpAffine(img, M2, (c,r))

flip = cv2.flip(img, 1)

p1 = np.float32([[50,50],[c-50,50],[50,r-50],[c-50,r-50]])
p2 = np.float32([[0,0],[c,0],[50,r],[c-50,r]])

M3 = cv2.getPerspectiveTransform(p1,p2)
pers = cv2.warpPerspective(img, M3, (c,r))

imgs = [img, resize, translate, rotate, flip, pers]
titles = ["Original","Resize","Translate","Rotate","Flip","Perspective"]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(imgs[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```
