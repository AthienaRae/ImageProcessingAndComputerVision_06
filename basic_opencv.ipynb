```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("basic.jpeg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

b,g,r = cv2.split(img)
merge = cv2.merge((b,g,r))

print("Shape:", img.shape)
print("Size:", img.size)
print("Type:", img.dtype)

plt.subplot(2,3,1)
plt.imshow(gray, cmap="gray")
plt.title("Gray")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(rgb)
plt.title("RGB")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(hsv)
plt.title("HSV")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(r, cmap="gray")
plt.title("Red Channel")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(g, cmap="gray")
plt.title("Green Channel")
plt.axis("off")

plt.subplot(2,3,6)
plt.imshow(b, cmap="gray")
plt.title("Blue Channel")
plt.axis("off")

plt.tight_layout()
plt.show()

cv2.imwrite("output.jpg", merge)
```
